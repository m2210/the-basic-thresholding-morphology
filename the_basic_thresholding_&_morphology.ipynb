```python
import numpy as np 
import matplotlib
from matplotlib import pyplot as plt 
import cv2
from google.colab.patches import cv2_imshow
```

```python
# upload image and make it binary in one step
img = cv2.imread('/content/image-3.jpeg',0)
cv2_imshow(img)
```

```python
img_a = np.asarray(img)
print(img_a)
plt.plot(img_a)
plt.show()
```

```python
#use threshold function to make image (whuite and black)
thresholding , re =cv2.threshold(img  , 61 , 255 , cv2.THRESH_BINARY)
cv2_imshow(re)
cv2.waitKey(0)
```

```python
#define kernal 
kernal = np.ones((3,3) ,np.uint8 )
#print(kernal)

# do erosion on our thresholding  image
erosion = cv2.erode(re , kernal, iterations=1)
#erosion = cv2.erode(re , kernal, iterations=2)
#print(erosion)
# show erosion effect on image 
cv2_imshow(erosion)

# see what happen if we change itteration 
```

```python
# see what happen if we change itteration 
erosion_2 = cv2.erode(re , kernal, iterations=4)
cv2_imshow(erosion_2)
```

```python
# define dillation 
dilation = cv2.dilate(erosion , kernal, iterations=1)
cv2_imshow(dilation)
```

```python
# see what happen if we change itteration 
dilation_2= cv2.dilate(erosion , kernal, iterations=4)
cv2_imshow(dilation_2)
```

```python
# define morphology 
starting=cv2.morphologyEx(re , cv2.MORPH_OPEN , kernal)
cv2_imshow(starting)
closing =cv2.morphologyEx(re,cv2.MORPH_CLOSE , kernal)
cv2_imshow(closing)
```
